```python
import pandas as guerreroDragon
import matplotlib.pyplot as meGraficas

tarjetasCredito = guerreroDragon.read_csv("/content/trx_tarjetas_credito.csv", sep=",")
```

```python
getGender = guerreroDragon.crosstab(index=tarjetasCredito["Gender"],columns="Gender")
print(getGender)

pieGender = getGender.loc[getGender.index]["Gender"]
meGraficas.pie(pieGender, autopct='%1.1f%%', shadow=True,
                   startangle=90, explode=(0.1, 0))
meGraficas.legend(getGender.index,
                  title="Generos",
                  loc="center left",
                  bbox_to_anchor=(1, 0, 0.5, 1))
meGraficas.title("Uso de tarjetas por genero")
meGraficas.rcParams['figure.facecolor'] = 'oldlace'
meGraficas.show

getGender.to_excel("Credit_Card_For_Gender.xlsx")
```

```python
getCityType = tarjetasCredito.groupby(["City", "Card Type"]).sum()
print(getCityType)

sepCityType = getCityType[:10]

pieCityType = sepCityType.loc[sepCityType.index]
meGraficas.pie(pieCityType, autopct='%1.1f%%', shadow=True, startangle=90,
               explode=(0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07, 0.07))
meGraficas.legend(pieCityType.index,
                  title="Ciudades y Tipos de tarjetas",
                  loc="center left",
                  bbox_to_anchor=(1, 0, 0.5, 1))
meGraficas.title("Tipo de tarjetas y ciudades")
meGraficas.rcParams['figure.facecolor'] = 'oldlace'
meGraficas.show

getCityType.to_excel("Card_Type_For_City.xlsx")
```

```python
getUse = guerreroDragon.crosstab(index=tarjetasCredito["Exp Type"],columns="Use")
print(getUse)

pieUse = getUse.loc[getUse.index]["Use"]
meGraficas.pie(pieUse, autopct='%1.1f%%', shadow=True, startangle=90,
               explode=(0.07, 0.07, 0.07, 0.07, 0.07 ,0.07))
meGraficas.legend(pieUse.index,
                  title="Gastos",
                  loc="center left",
                  bbox_to_anchor=(1, 0, 0.5, 1))
meGraficas.title("Uso de tarjetas por gasto")
meGraficas.rcParams['figure.facecolor'] = 'oldlace'
meGraficas.show

getUse.to_excel("Credit_Card_For_Use.xlsx")
```
